```
import numpy as np
import matplotlib.pyplot as plt #파이썬 표준 그래프 패키지
```

```
#파이썬 리스트
a = [1, 2]
b = [3, 4]

print(a+b)
```

```
#넘파이 배열
a = np.array([1, 2])
b = np.array([3, 4])

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```
#y=x^2
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

```
#y=x^2
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.show()
```

```
plt.scatter(x, y)
plt.show()
```

```
x = np.random.randn(1000) #랜덤 넘버 1000개 생성
y = np.random.randn(1000) #랜덤 넘버 1000개 생성
plt.scatter(x, y)
plt.show()
```

```
x1 = np.random.randn(1000) - 2
x2 = np.random.randn(1000) + 2
y = np.random.randn(1000)

plt.scatter(x1, y)
plt.scatter(x2, y)
plt.show()
```
